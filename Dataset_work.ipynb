```python
import pandas as pd
from bs4 import BeautifulSoup

# Load your StackOverflow dump data into a DataFrame
df = pd.read_csv('Data.csv')

# Function to extract body, code, and images from the body text
def extract_body_code_images(body):
    soup = BeautifulSoup(body, 'html.parser')

    # Extract the body text
    body_text = soup.get_text()

    # Extract code blocks
    code_blocks = soup.find_all('code')
    code_text = '\n'.join([code.get_text() for code in code_blocks])

    # Extract image URLs
    img_tags = soup.find_all('img')
    img_urls = [img['src'] for img in img_tags if 'src' in img.attrs]

    return body_text, code_text, img_urls

# Apply the extract_body_code_images function to the 'Body' column
extracted_data = df['Body'].apply(extract_body_code_images)

# Create new DataFrame columns
df['Body'] = extracted_data.apply(lambda x: x[0])  # Replace the 'Body' column
df['CodeText'] = extracted_data.apply(lambda x: x[1])
df['ImageURLs'] = extracted_data.apply(lambda x: x[2])  # Store image URLs as an array

# Save the updated DataFrame to the same CSV file
df.to_csv('Data.csv', index=False)
```
